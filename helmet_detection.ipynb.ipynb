```python
import zipfile

with zipfile.ZipFile("/content/Helmet Detection 3.v3i.yolov8.zip", 'r') as zip_ref:
    zip_ref.extractall("/content")
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  # fastest model

model.train(
    data="/content/data.yaml",
    epochs=5,        # fast but decent
    imgsz=416,       # smaller = faster
    batch=16,        # good for GPU
    device=0         # use GPU
)
```

```python
from ultralytics import YOLO

model = YOLO("/content/runs/detect/train/weights/best.pt")

model.predict(
    source="/content/test/images",
    save=True
)
```
